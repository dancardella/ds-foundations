```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv", sep = ",")

recent_grads.head()
recent_grads.tail()
recent_grads.dropna(how="any")
```

Create a for loop that plots the histogram of sample size, median, employed, full time, unemployment rate, men and women (remember to use the exact column titles)

```python
wanted = ('Sample_size', 'Employed', 'Full_time','Unemployment_rate', 'Median', 'Men', 'Women')
sub_list_wanted = ('Employed', 'Full_time', 'Men', 'Women')
for stat in wanted:    
    fig = plt.figure()
    ax = plt.subplot(111)
    recent_grads.plot(y=stat,ax=ax, kind='hist')
    if stat in sub_list_wanted:    
        ax.set_xlim(0,150000)
```

Make a bar plot of majors vs "ShareWomen" (you should run different slices since there are many different majors)

```python
# A conjunction query using two columns >>> df[(df['rbi'] > 10) & (df['team'] == 'TOR')] 

rank1= recent_grads[recent_grads['Rank'] < 45]
rank2= recent_grads[(recent_grads['Rank'] > 45) & (recent_grads['Rank'] < 90)]
rank3= recent_grads[(recent_grads['Rank'] > 90) & (recent_grads['Rank'] < 135)]
rank4= recent_grads[(recent_grads['Rank'] > 135)]
```

```python
SIZE = 38
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

#plt.rc('font', size=SIZE)                # controls default text sizes
#plt.rc('axes', titlesize=SIZE)           # fontsize of the axes title
#plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SIZE)          # fontsize of the tick labels
plt.rc('ytick', labelsize=SIZE)          # fontsize of the tick labels
plt.rc('legend', fontsize=SIZE)          # legend fontsize
#plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
%matplotlib inline
fig = plt.figure()
ax = plt.subplot(111)
rank1.plot(x='Major',y='ShareWomen',ax=ax, kind='bar', figsize=(55,30))
plt.xlabel('Major', fontsize=108)
fig.suptitle('Rank 1 Salaries - Share Women', fontsize=60)
```

```python
%matplotlib inline
fig = plt.figure()
ax = plt.subplot(111)
rank2.plot(x='Major',y='ShareWomen',ax=ax, kind='bar', figsize=(55,30))
plt.xlabel('Major', fontsize=108)
fig.suptitle('Rank 2 Salaries - Share Women', fontsize=60)
```

```python
 %matplotlib inline
fig = plt.figure()
ax = plt.subplot(111)
rank3.plot(x='Major',y='ShareWomen',ax=ax, kind='bar', figsize=(55,30))
plt.xlabel('Major', fontsize=108)
fig.suptitle('Rank 3 Salaries - Share Women', fontsize=60)
```
